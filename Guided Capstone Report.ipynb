## Guided Capstone Report

### Overview

Big Mountain Resort is a ski resort locating at Big Mountain that sells services and products to skiers and snowboarders. Recently, after spending 1.5 million dollars on additional chair lift, the company would like to implement a data driven investment strategy by either reducing operating cost or raising the ticket price.

In this project, I analyze the dataset of all ski resorts across the country and project the ticket price for Big Mountain Resort based on their current features. Some investment strategies were considered for the resort to raise their ticket price based on the same model. Ultimately, I concluded that compared to all other resorts in the United States, Big Mountain Resort should raise their ticket price by \\$13.22 and further increase of ticket price can implemented if they improve certain features of the resort.

### Modeling

The models I implemented in this project are linear regression and random forest. Their best hyperparameters, number of predictors for LR and number of trees for RF, were selected through 5-fold cross-validation. The comparison result is shown in the table below:

| Models | Train mean MAE | Train Sd | Test MAE |
| :- | -: | :-: | -: |

### Prediction and Conclusion

Running through our model, it predicted a ticket price of **\\$94.22** based on the features in Big Mountain Resort. This is an increase of **\\$13.22** from the current price. 

Based on the features importance graph above, some investment/cost reduction scenarios were proposed:
1. Closing down the 10 least popular runs in the resort
2. Adding a run, raise the vertical drop by 150ft, and installing an addition chair lift
3. Adding 2 more acres of snow making on top of scenario 2
4. Increase the longest run by 0.2 mile and adding 4 acres of snow making
